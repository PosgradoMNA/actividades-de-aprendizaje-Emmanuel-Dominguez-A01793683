7. Reduce las dimensiones con PCA, si consideras necesario.


*   Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70% 
*   Indica la importancia de las variables en cada componente




```python
# creamos un data frame con los valores escalados para visualizar mejor la
# información:

scaled_df = pd.DataFrame(scaled, columns=imp_df.columns)
scaled_df.head()
```

```python
from sklearn.decomposition import PCA
pcs = PCA()

pcs_t = pcs.fit_transform(scaled_df)
pcs_t
```

```python
# Calculamos el % de la varianza explicada por cada componente del PCA, así como
# El acumulado de la varianza explicada por cada componente

pcsSummary_df = pd.DataFrame({'Desviación estándar': 
                              np.sqrt(pcs.explained_variance_).round(4),
                              'Varianza explicada': 
                              pcs.explained_variance_ratio_.round(4)*100,
                              'Varianza acumulada': 
                              np.cumsum(pcs.explained_variance_ratio_).round(4)*100})

pcsSummary_df
```

```python
# Creamos una lista con los nombres de cada componente: PC1, PC2 ... PC13

pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df.columns))]
pcsSummary_df.index = pcs_labels
pcsSummary_df
```

```python
# Creamos un dataframe con los valores o peso de cada componente.

pcs_df = pd.DataFrame(pcs_t, columns =pcs_labels)
print("Varianza total variables originales: ", scaled_df.var().sum().round(2))
print("Varianza total de los componentes: ", pcs_df.var().sum().round(2))
```
